### Data Analysis and Visualizations

```python
# Show Reservation Status
is_canceled = df['is_canceled'].value_counts(normalize=True)
print(is_canceled)

plt.figure(figsize=(7,4))
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='black')
plt.title('Reservation Status')
```

```python
a = sns.countplot(data= df, x='hotel', hue='is_canceled', palette='Blues')
plt.title("Reservation Status in different hotels")
plt.xlabel('Hotel')
plt.ylabel('No. of reservations')
plt.legend(['Not canceled','canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15,7))
plt.title('Average Daily Rate in Resort and City Hotel',fontsize=25)
plt.plot(resort_hotel.index,resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=15)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
x = sns.countplot(data=df, x='month', hue='is_canceled')
plt.title('Reservation Status per month')
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```
